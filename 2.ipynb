```python
import matplotlib.pyplot as plt
import numpy as np
import random
import pandas as pd
from scipy import integrate
import cmath
from matplotlib.ticker import ScalarFormatter
```

### a)

```python

# Parmetros  
beta = 0.05  
gamma = 0.1  
N = 1000000  
S = 700000  
I = 300000  
R = 0  
delta_t = 0.01  
t_max = 500  

# Arrays para almacenar los valores a graficar  
tiempo = int(t_max / delta_t)  
S_v = np.zeros(tiempo)  
I_v = np.zeros(tiempo)  
R_v = np.zeros(tiempo)  

# Simulacin  
for n in range(tiempo):  
    S_v[n] = S  
    I_v[n] = I  
    R_v[n] = R  
    
    S_nuevo = S - delta_t * (beta * I * S / N)  
    I_nuevo = I + delta_t * (beta * I * S / N - gamma * I)  
    R_nuevo = R + delta_t * gamma * I  
    
    S, I, R = S_nuevo, I_nuevo, R_nuevo 

# Graficar los resultados  
t = np.linspace(0, t_max, tiempo)  
plt.figure(figsize=(10, 6))  
plt.plot(t, S_v, color="deeppink", label='Susceptibles (S)')  
plt.plot(t, I_v, color="lightskyblue", label='Infectados (I)')  
plt.plot(t, R_v, color="cadetblue", label='Recuperados (R)')  
plt.xlabel('Tiempo')  
plt.ylabel('Nmero de individuos')  
plt.title('Modelo SIR: Simulacin de Epidemia')  
plt.legend()  
plt.grid()  
plt.show()
```

```python

# Parmetros  
N = 1000000  
I0 = 100      
f_vac = 0.5  
R0 = f_vac * N  
S0 = N - I0 - R0   
beta = 0.2    
gamma = 0.04  


t = np.linspace(0, 500, 500)


S = np.zeros(len(t))  
I = np.zeros(len(t))  
R = np.zeros(len(t))  


S[0] = S0  
I[0] = I0  
R[0] = R0  

# Simulacin de SIR  
for day in range(1, len(t)):  
    S[day] = S[day-1] - beta * S[day-1] * I[day-1] / N  
    I[day] = I[day-1] + beta * S[day-1] * I[day-1] / N - gamma * I[day-1]  
    R[day] = R[day-1] + gamma * I[day-1]  

# Fraccin no infectada al final  
no_infectados = S[-1] + R[-1]  # Susceptibles + Recuperados  

# Resultados  
print(f'Nmero de individuos no infectados al final: {no_infectados:.0f}')
```
