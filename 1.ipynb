```python
import matplotlib.pyplot as plt
import numpy as np
import random
import pandas as pd
from scipy import integrate
import cmath
from matplotlib.ticker import ScalarFormatter
```

```python
def mapeologis(r,x):
    funcion = r*x*(1-x)
    return funcion
```

### a)

```python
tiempo = np.linspace(0,1)

plt.figure()
plt.plot(tiempo, tiempo, "-or", color="violet", label = r"$f(x) = x$")
plt.plot(tiempo,mapeologis(0.5,tiempo),"-xb", color="green", label = r"$f(x) = r  \cdot x(1-x)$")
plt.legend()
plt.title("Gráfica con r = 0.5", fontweight='bold')
plt.xlabel("x(t)")
plt.ylabel("x(t+1)")
plt.show()
```

```python


r = 0.5 
x0 = 0.1  
n_i = 100    


x_v = np.zeros(n_i)  
x_v[0] = x0  


for n in range(1, n_i):  
    x_v[n] = r * x_v[n - 1] * (1 - x_v[n - 1])  


plt.figure(figsize=(10, 5))  
plt.plot(x_v, marker='o', linestyle='-')  
plt.title(f'Evolución de x para r = {r} (0 < r < 1)')  
plt.xlabel('Iteraciones')  
plt.ylabel('Valor de x')  
plt.ylim(0, 0.5)  
plt.grid()  
plt.axhline(0, color='black', lw=0.5, ls='--')  
plt.axhline(0.5, color='red', lw=0.5, ls='--', label='Límite superior')  
plt.legend()  
plt.show()
```

```python

# Parámetros de configuración  
r_values = np.linspace(2, 8, 100)  # 100 valores de r entre 2 y 8  
n_iterations = 100000  # Número total de iteraciones  
last_iterations = 100  # Últimos 100 valores a graficar  

# Configurar el gráfico  
plt.figure(figsize=(12, 8))  
plt.title("Diagrama de Bifurcación del Mapeo Logístico")  
plt.xlabel("r")  
plt.ylabel("x")  
plt.ylim(-0.1, 1.1)  # Limitar el eje y para mejor visualización  

# Calcular el diagrama de bifurcación  
for r in r_values:  
    # Inicializar un valor inicial aleatorio en [0, 1]  
    x = np.random.rand()  

    # Realizar 1000 iteraciones  
    for _ in range(n_iterations):  
        x = r * x * (1 - x)  # Aplicar el mapeo logístico  

        # Guardar los últimos 100 valores  
        if _ >= n_iterations - last_iterations:  
            plt.plot(r, x, 'bo', markersize=0.5)  # Graficar el punto  

# Mostrar el gráfico  
plt.grid()  
plt.show()
```
